# 3. Formación de muestras de entrada a la red

- Se realiza la construcción de las **muestras de entrada** a la red
- Cada **muestra** está compuesta de 2 batimetrías pasadas **(t y t-1)**
- El **target** es la batimetría en **(t+1)**

En este caso, cada muestra utiliza **dos meses consecutivos** como predictor del **mes siguiente**

## 3.1. Pasos

1. Se cargan las teselas construidas anteriormente 
2. Se cargan los metadatos temporales
3. Se construyen ventanas temporales deslizantes por teselas
4. Se generan tensores de entrada y salida

Genera:
- **`X`**: tensor con las entradas temporales
- **`Y`**: tensor con el target espacial correspondiente al tiempo siguiente




## 3.2. Parámetros principales

El comportamiento del script está controlado por tres parámetros base:


```python
SECTOR = "X6By5"
TILE_SIZE = 68
TIME_STEP = 2  # (t-1, t) -> t+1
```

## 3.3. Construcción de muestras supervisadas

```python
X_list = []
Y_list = []
meta_list = []

for tile_id, g in df_tiles.groupby("tile_id"):
    idxs = g.index.to_numpy()

    for k in range(len(idxs) - TIME_STEP):
        idx_input = idxs[k : k + TIME_STEP]
        idx_target = idxs[k + TIME_STEP]

        x_sample = X_tiles[idx_input]   # (TIME_STEP, 68, 68)
        y_sample = X_tiles[idx_target]  # (68, 68)

        X_list.append(x_sample)
        Y_list.append(y_sample)
```

## 3.4. Forma final de los datos

Al finalizar, los tensores quedan con la siguiente estructura conceptual:

- **`X`**: `(n_samples, TIME_STEP, TILE_SIZE, TILE_SIZE, 1)`
- **`Y`**: `(n_samples, TILE_SIZE, TILE_SIZE, 1)`

Esto significa que cada muestra de entrada contiene una pequeña secuencia temporal de mapas espaciales, y cada salida corresponde al mapa espacial del tiempo siguiente.
